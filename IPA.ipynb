```python
# import library
import pandas as pd
import numpy as np
# split train set dan test set
from sklearn.model_selection import train_test_split

# standardization
from sklearn.preprocessing import StandardScaler

# model Decision Tree
from sklearn.ensemble import RandomForestClassifier

# menilai performa model
from sklearn.metrics import classification_report

# save model untuk deploy
import pickle

# library untuk mengabaikan pesan warning
import warnings
warnings.filterwarnings("ignore")
```

PREDIKSI IPA

```python
df = pd.read_csv("datafiks.csv")
df.head()
```

```python
#Mencoba Rekomendasi Data IPA

ipa = df[['TipeStudi', 'MTK [IPA]', 'Indo [IPA]', 'Eng [IPA]', 'Biologi', 'Kimia', 'Fisika']]
ipa.dropna(inplace=True)
ipa
```

```python
ipa.isnull().sum()
```

```python
ipa["MTK [IPA]"] = ipa["MTK [IPA]"].astype('category')
ipa["MTK_IPA"] = ipa["MTK [IPA]"].cat.codes
ipa = ipa.drop(['MTK [IPA]'],axis = 1)

ipa["Indo [IPA]"] = ipa["Indo [IPA]"].astype('category')
ipa["Indo_IPA"] = ipa["Indo [IPA]"].cat.codes
ipa = ipa.drop(['Indo [IPA]'],axis = 1)

ipa["Eng [IPA]"] = ipa["Eng [IPA]"].astype('category')
ipa["Eng_IPA"] = ipa["Eng [IPA]"].cat.codes
ipa = ipa.drop(['Eng [IPA]'],axis = 1)

ipa["Biologi"] = ipa["Biologi"].astype('category')
ipa["Biologi_IPA"] = ipa["Biologi"].cat.codes
ipa = ipa.drop(['Biologi'],axis = 1)

ipa["Kimia"] = ipa["Kimia"].astype('category')
ipa["Kimia_IPA"] = ipa["Kimia"].cat.codes
ipa = ipa.drop(['Kimia'],axis = 1)

ipa["Fisika"] = ipa["Fisika"].astype('category')
ipa["Fisika_IPA"] = ipa["Fisika"].cat.codes
ipa = ipa.drop(['Fisika'],axis = 1)
```

```python
ipa
```

```python
X = ipa.drop(['TipeStudi'],axis=1).values
y = ipa['TipeStudi'].values
```

```python
X
```

```python
from imblearn.over_sampling import RandomOverSampler
ovr = RandomOverSampler(random_state=42)
X_res, y_res = ovr.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(bootstrap=False, ccp_alpha=0.0,
                       class_weight='balanced_subsample', criterion='gini',
                       max_depth=6, max_features=1.0, max_leaf_nodes=None,
                       max_samples=None, min_impurity_decrease=0,
                       min_samples_leaf=4,
                       min_samples_split=7, min_weight_fraction_leaf=0.0,
                       n_estimators=300, n_jobs=-1, oob_score=False,
                       random_state=123, verbose=0, warm_start=False)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
def Recommender(X):
    a = df_rec[df_rec['TipeStudi']==X].values
    b = a[:,0]
    print(f'Rekomendasi Jurusan: {b}', end='')
```

```python
def MTK(X):
  if X == 'Sedang':
    result = 2
  elif X == 'Sangat Tinggi':
    result = 1
  elif X == 'Tinggi':
    result = 3
  elif X == 'Rendah' or X == 'Sangat Rendah' :
    result = 0
  return result

def KIM(X):
  if X == 'Sangat Rendah':
    result = 1
  elif X == 'Sangat Tinggi':
    result = 2
  elif X == 'Sedang':
    result = 3
  elif X == 'Tinggi':
    result = 4
  elif X == 'Rendah':
    result = 0
  return result

def FIS(X):
  if X == 'Sangat Tinggi':
    result = 1
  elif X == 'Sedang':
    result = 2
  elif X == 'Tinggi':
    result = 4
  elif X == 'Rendah' or X == 'Sangat Rendah' :
    result = 0
  return result
```

```python
MTK_IPA = 'Rendah' #@param ['Sedang','Sangat Tinggi','Tinggi','Rendah','Sangat Rendah','Rendah Sekali']{type:"string"}
MTK_IPA = MTK(MTK_IPA)
Indo_IPA = 'Sangat Tinggi' #@param ['Sedang','Sangat Tinggi','Tinggi','Rendah','Sangat Rendah','Rendah Sekali']{type:"string"}
Indo_IPA = MTK(Indo_IPA)
Eng_IPA = 'Sangat Tinggi' #@param ['Sedang','Sangat Tinggi','Tinggi','Rendah','Sangat Rendah','Rendah Sekali']{type:"string"}
Eng_IPA = MTK(Eng_IPA)
Biologi_IPA = 'Sangat Rendah' #@param ['Sedang','Sangat Tinggi','Tinggi','Rendah','Sangat Rendah','Rendah Sekali']{type:"string"}
Biologi_IPA = MTK(Biologi_IPA)
Kimia_IPA = 'Rendah' #@param ['Sedang','Sangat Tinggi','Tinggi','Rendah','Sangat Rendah','Rendah Sekali']{type:"string"}
Kimia_IPA = KIM(Kimia_IPA)
Fisika_IPA = 'Tinggi' #@param ['Sedang','Sangat Tinggi','Tinggi','Rendah','Sangat Rendah','Rendah Sekali']{type:"string"}
Fisika_IPA = FIS(Fisika_IPA)

# initialize list of lists
data = [[MTK_IPA,Indo_IPA,Eng_IPA,Biologi_IPA,Kimia_IPA,Fisika_IPA]]

# Create the pandas DataFrame
New_Data = pd.DataFrame(data, columns=['MTK_IPA','Indo_IPA','Eng_IPA','Biologi_IPA','Kimia_IPA','Fisika_IPA'])

Prediction = model.predict(New_Data)
pred = Prediction
rec_str = pred.item()[:]

print("\n")
print(f'Program Studi Yang Direkomendasikan: {rec_str}', end='')
print("\n")
rec = Recommender(rec_str)
```
